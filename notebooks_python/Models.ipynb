```python
###---- Librairies 
import numpy as np 
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, roc_auc_score, f1_score
```

```python
###--- Importation data 
cardio = pd.read_csv("cardio_impute.csv")

cardio.head()
```

```python
# Dimension 
cardio.shape
```

```python
###--- Info data 
cardio.info()
```

```python
###--- Describe variables  
cardio.describe().T
```

```python
cardio["cardio"].value_counts(normalize=True)
```

```python
###--- Sélection de variables 
cardio_select = cardio[["age", "height", "weight", "imc", "ap_hi", "ap_lo", "cardio"]]

###--- mean by cardio 
cardio_select.groupby("cardio").mean()
```

```python
###--- Fonction pour convertir le type de certaines variables 

def convert_to_category(df, max_unique=3):
    for col in df.columns:
        if df[col].dtype in ['int64', 'float64']:  # numérique au départ
            if df[col].nunique() <= max_unique:
                df[col] = df[col].astype('category')
    return df

###--- convertir 
data_final = convert_to_category(cardio)

###--- verification 
data_final.dtypes.value_counts()
```

```python
###--- Split data 
X = data_final[["age", "ap_hi", "ap_lo", "imc", "hypertendu", "cholesterol"]]
y = data_final["cardio"]
```

```python
###--- train an test set 
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size =0.2, stratify=y, random_state = 234)
```

```python
###--- Shape of train and test set 
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

### **REGRESSION LOGISTIQUE**

```python
from sklearn.model_selection import cross_val_score, GridSearchCV
```

```python
parameters = {
    "penalty" : ["l1", "l2"], 
    "solver" : ["lbfgs", "sag", "saga"],
    "max_iter" : np.array([100, 200])
}


grid = GridSearchCV(LogisticRegression(), parameters, cv=5)

grid.fit(X_train, y_train)
```

```python
model_reg = grid.best_estimator_

model_reg
```

```python
###--- fit model 
# reg_log = LogisticRegression()
model = model_reg.fit(X_train, y_train)
```

```python
###--- predict on train set 
X_train_pred = model.predict(X_train)

###--- metrics on train set 
accuracy_training = accuracy_score(X_train_pred, y_train)
recall_training = recall_score(X_train_pred, y_train)
precision_training = precision_score(X_train_pred, y_train)
roc_auc_training = roc_auc_score(X_train_pred, y_train)
f1_score_training = f1_score(X_train_pred, y_train)

###--- accuracy and precsion on train set 
print("accuracy_trainset  :", accuracy_training)
print("recall_training    :", recall_training)
print("precision_trainset :", precision_training)
print("roc_auc_training   :", roc_auc_training)
print("f1_score_training  :", f1_score_training)

```

```python
###--- Predict on test set 
X_test_pred = model.predict(X_test)

###--- metrics on test set 
accuracy_test = accuracy_score(X_test_pred, y_test)
recall_test = recall_score(X_test_pred, y_test)
precision_test = precision_score(X_test_pred, y_test)
roc_auc_test = roc_auc_score(X_test_pred, y_test)
f1_score_test = f1_score(X_test_pred, y_test)

###--- accuracy and precsion on test set 
print("accuracy_test  :", accuracy_test)
print("recall_test    :", recall_test)
print("precision_test :", precision_test)
print("roc_auc_test   :", roc_auc_test)
print("f1_score_test  :", f1_score_test)
```

### **GradientBoostingClassifier**

```python
from sklearn.ensemble import GradientBoostingClassifier

# Modèle
gb_clf = GradientBoostingClassifier(
    learning_rate=0.05,
    n_estimators=300,
    max_depth=3,
    subsample=0.8,
    random_state=42
)

model_gb = gb_clf.fit(X_train, y_train)

# Prédictions
gb_train_pred = model_gb.predict(X_train)
gb_test_pred = model_gb.predict(X_test)

# Scores Training
metrics_train = {
    "Accuracy": accuracy_score(y_train, gb_train_pred),
    "Recall": recall_score(y_train, gb_train_pred),
    "Precision": precision_score(y_train, gb_train_pred),
    "ROC AUC": roc_auc_score(y_train, gb_train_pred),
    "F1 Score": f1_score(y_train, gb_train_pred)
}

# Scores Test
metrics_test = {
    "Accuracy": accuracy_score(y_test, gb_test_pred),
    "Recall": recall_score(y_test, gb_test_pred),
    "Precision": precision_score(y_test, gb_test_pred),
    "ROC AUC": roc_auc_score(y_test, gb_test_pred),
    "F1 Score": f1_score(y_test, gb_test_pred)
}

print("===== Performance Train =====")
for k, v in metrics_train.items():
    print(f"{k:10s}: {v:.4f}")

print("\n===== Performance Test =====")
for k, v in metrics_test.items():
    print(f"{k:10s}: {v:.4f}")
```

```python

from sklearn.metrics import classification_report

# Définition du modèle
gb = GradientBoostingClassifier(random_state=42)

# Grille des hyperparamètres à tester
param_grid = {
    "n_estimators": [100, 200, 300],
    "learning_rate": [0.01, 0.05, 0.1],
    "max_depth": [2, 3, 4],
    "subsample": [0.7, 0.8, 1.0],
    "max_features": ["sqrt", "log2"]
}

# Grid Search avec validation croisée
grid_search = GridSearchCV(
    estimator=gb,
    param_grid=param_grid,
    scoring="f1",         # optimise le F1 score
    cv=5,                 # Cross-validation
    n_jobs=-1,            # Utilisation de tous les CPU
    verbose=2
)

grid_search.fit(X_train, y_train)

# Meilleurs hyperparamètres
print("\n✅ Best Parameters:")
print(grid_search.best_params_)

# Meilleur modèle optimisé
best_gb = grid_search.best_estimator_

# Évaluation sur les données test
y_pred_test = best_gb.predict(X_test)

print("\n===== Best Model Performance on Test Set =====")
print(classification_report(y_test, y_pred_test))
```
